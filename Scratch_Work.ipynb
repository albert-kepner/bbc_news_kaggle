```python
def pick_fraction_df(df, fraction=0.1):
    len1 = len(df)
    sample_size = round(fraction * len1)
    indices = random.sample(range(len1), sample_size)
    return df.iloc[indices,:]

pick_fraction_df(df)
```

```python
def pick_fraction(array_like, fraction=0.1):
    array_like = np.array(array_like)
    len1 = len(array_like)
    sample_size = round(fraction * len1)
    indices = random.sample(range(len1), sample_size)
    remainder_indices = list(set(range(len1)) - set(indices))
    sample, remainder = array_like[indices], array_like[remainder_indices]
    return sample, remainder

pick_fraction(np.array(range(40)), 0.1)
```
